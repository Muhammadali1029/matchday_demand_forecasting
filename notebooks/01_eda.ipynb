```python
df_1_filtered = df_1[['ORDER_HEADER_ID', 'ITEM_DESCRIPTION', 'QTY', 'SUB_TOTAL', 'PAID']]
df_1_filtered.head(15)
```

consolidate the rows by ODER_HEADER_ID

```python
agg_funcs = {
    'ITEM_DESCRIPTION': lambda x: ', '.join(x.dropna()),
    'QTY': 'sum',
    'SUB_TOTAL': 'sum',
    'PAID': 'first'
}

df_1_consolidated = df_1_filtered.groupby('ORDER_HEADER_ID').agg(agg_funcs).reset_index()
print(df_1_consolidated.shape)
df_1_consolidated.head()
```

SUBTOTAL string and seperated by commas so first need to replace commas with point and then turn to numeric. 

```python
df_1_filtered['SUB_TOTAL'] = df_1_filtered['SUB_TOTAL'].str.replace(',', '.', regex=True).astype(float)
df_1_filtered.head()
```

Now the consolidation should work perfectly

```python
agg_funcs = {
    'ITEM_DESCRIPTION': lambda x: ', '.join(x.dropna()),
    'QTY': 'sum',
    'SUB_TOTAL': 'sum',
    'PAID': 'first'
}

df_1_consolidated = df_1_filtered.groupby('ORDER_HEADER_ID').agg(agg_funcs).reset_index()
print(df_1_consolidated.shape)
df_1_consolidated.head(20)
```

```python
df_2_Total = df_2['TOTAL'].str.replace(',', '.', regex=True).astype(float)

mismatched_row = df_1_consolidated['SUB_TOTAL'] != df_2_Total
pd.concat([df_1_consolidated[mismatched_row], df_2_Total[mismatched_row]], axis=1)
```
